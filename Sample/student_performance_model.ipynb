```python
print('Topic: Student Performace Classifier\n')

print('Step1: Importing the libraries\n')

import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

print('Step2: Loading the dataset\n')

df = pd.read_csv('student_performance_dataset.csv')

print('Step3: Data preparation\n')

encoder = OrdinalEncoder()

print(df.Performance_Level.unique(), '\n')

df['Performance_Level'] = encoder.fit_transform(df[['Performance_Level']])

print(df.Performance_Level.unique(), '\n')

X = df.iloc[:, :-1]
y = df.iloc[:, -1]

print('Step4: Splitting the dataset\n')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

print('Step5: Model creation\n')

model = LogisticRegression()

print('Step6: Model training\n')

model.fit(X_train, y_train)

print('Step spl: Model score')

print('Model training score: ', model.score(X_train, y_train))

print('Model testing score: ', model.score(X_test, y_test), '\n')

y_predicted = model.predict(X_test[:1])

d = {0.0: 'Average Performer', 1.0: 'High performer', 2.0: 'Needs Improvement'}

d[y_predicted[0]]
```
